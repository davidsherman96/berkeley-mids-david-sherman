# Assessing The Value of AGM's Customers

The first step in determining who AGM's best customers are entails figuring out who has been supporting the business lately. Assuming we're reviewing the data in today or in the near future, AGM should place more value on customers who have purchased an AGM meal more recently as they would be more likely to make a repeat purchase in the near future, as compared to someone who made a purchase last month. In order to determine who those customers are, we would need to utilize the (1) "sales" data table and (2) "customer" data table. The "sales" table possesses data related to which dates purchases were made and who made those orders - we could utilize the "sale_date" field to determine the most recent purchases then tie those orders to the "customer_id" field. We could create a "recency_rank" field that ranks each purchase based on the sale_date, assigning a better rank (i.e. 1) to the most recent date. Once we pull the customer_ids for the more recent buyers, we could pull in their contact information from the "customers" table by joining on customer_id. 

The next step involves understanding which customers are making the most repeat purchases. Similar to understanding who has been making more recent purchases, we can leverage the (1) "sales" and (2) "customer" data tables to determine who are the customers that are making the most purchases. Using the "sales" table, we can count the total number of sale_ids (count(sale_id)) for each customer_id to count the total number of purchases per customer. We could then develop a "frequency_rank" field that ranks each customer based on the total number of purchases they make. Once we pull the customer_ids for the more frequent buyers, we could pull in their contact information from the "customers" table by joining on customer_id. 

AGM would also want to know how much are certain customers spending on their meals. Again, we can use the (1) "sales" and (2) "customer" data tables to determine who are the customers that are spending the most on AGM meals. Using the "sales" table, we can sum the "total_amount" (sum(total_amount)) for each customer_id to quantify the total amount of money each customer has spent on AGM meals. We could then develop a "money_rank" field that ranks each customer based on the total amount of money they spend on AGM meals. Once we pull the customer_ids for the higher spenders, we could pull in their contact information from the "customers" table by joining on customer_id.

After conducting the above three analyses, we've now obtained three rank fields for each customer_id - "recency_rank", "frequency_rank", and "money_rank". To review, the lower the number, the more valuable the customer. For example, the customer with a "money_rank" value of 1 spent the most on AGM meals. The next step would be to weight each of the three ranked fields based on their importance to AGM. In my opinion, I think the weighting should be "recency_rank" as 20%, "frequency_rank" as 40%, and "money_rank" as 40%. The rationale for "recency_rank" being the lowest valued field is that even though we would want to prioritize more recent customers, we aren't 100% certain that they would make a repeat purchase over someone who made a purchase last month. That being said, number of purchases ("frequency_rank") and total amount spent ("money_rank") would tell a better story, hence which is why they are weighted more than the "recency_rank". We would then multiply each of the three ranks by their respective weights to arrive at a specific raw score for each customer_id. The lower the raw score (i.e. 1 would be the best), the more valuable/better that customer would be.
