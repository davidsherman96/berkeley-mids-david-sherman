```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

accuracies = {'Baseline':0.87, 
              'Logistic Regression':0.94, 
              'K-Nearest Neighbors':0.94,
              'Decision Tree Classifier':0.92,
              'Random Forest':0.95,
              'Adaboost':0.95,
              'Gradient Boosting':0.96,
              'Voting Hard':0.96,
              'Voting Soft':0.95
             }

accuracies_df = pd.DataFrame(list(accuracies.items()), columns=['Model', 'Accuracy'])
accuracies_df.set_index('Model', inplace=True) 

plt.figure(figsize=(14, 10))  # Set the figure size (optional)
accuracies_df['Accuracy'].plot(kind='bar', color='green', edgecolor='black')

def format_percent(x, pos):
    return f"{int(x*100)}%"

y_formatter = FuncFormatter(format_percent)
plt.gca().yaxis.set_major_formatter(y_formatter)

# Set plot labels and title
plt.xlabel('Model', fontsize=20)
plt.ylabel('Accuracy', fontsize=20)
plt.title('Model Accuracies', fontsize=20)

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
Model = ['Baseline','Logistic Regression','K-Nearest Neighbors','Decision Tree Classifier','Random Forest','Adaboost','Gradient Boosting','Voting Hard','Voting Soft']
no_diabetes_recall = [1, 0.99, 0.99, 0.95, 0.99, 1, 1, 1, 0.99]
diabetes_recall = [0, 0.60, 0.64, 0.74, 0.69, 0.69, 0.68, 0.68, 0.69]

x = np.arange(len(Model))  
width = 0.35

fig, ax = plt.subplots(figsize=(19, 10))

rects1 = ax.bar(x - width/2, no_diabetes_recall, width, label='No Diabetes', color='g')
rects2 = ax.bar(x + width/2, diabetes_recall, width, label='Diabetes', color='b')

ax.set_ylabel('Recall',fontsize=20)
ax.set_title('Recall by Model and Diabetes Diagnosis',fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(Model, rotation=45, ha='right',fontsize=15)
ax.legend()

def format_percent(x, pos):
    return '{:.0%}'.format(x)

y_formatter = FuncFormatter(format_percent)
ax.yaxis.set_major_formatter(y_formatter)

# Display values on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.0%}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12)

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
